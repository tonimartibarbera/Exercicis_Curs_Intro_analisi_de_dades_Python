```python
# A partir de l'exercici que ens demanem, podem decidir eliminar certes columnes.
# Pero primero, solucionemos el tema de los NaN
# Miramos cuántos NO son retraso >> criterio que el Delay sea menor de 15 minutos
df[df['ArrDelay']<15].shape[0]
```

```python
#Ahora ya tenemos a qué corresponden los NaN, además de los retrasos de menos de 15 minutos, que tampoco consideramos retraso
df.ArrDelay[df['ArrDelay']<15]=0
df
```

```python
# Vamos con los Cancelados y Desviados (los que no fueron cancelados ni desviados)
df = df[(df.Cancelled == 0) & (df.Diverted == 0)]
```

```python
# Ya no tenemos NaN
df.isnull().sum()
```

```python
# Miramos si hay du
df.duplicated().sum()
```

```python
# NO hay duplicados. Si hubiera, para eliminarlos: df.drop_duplicates()
# Elegimos las columnas que vamos a necesitar en el análisis
df_limpio = df[['FlightNum', 'AirTime','ArrDelay','UniqueCarrier']]
df_limpio.head()
```

```python
#Ahora hay que tomar la muestra de 200.000 registros. Lo haremos con la función sample, que lo hace aleatoriamente. 
# Para evitar que cada ejecución genere una sample diferente, le damos una valor al parámetro random_state
df_sample = df_limpio.sample(n=200000, random_state=12)
df_sample
```

```python
# Mostra quina és les aerolínies amb més endarreriments acumulats.
# Se puedes interpretar de varias forma: tiempo medio de retraso, tiempo acumulado de retrasos, número de retrasos
df_sample.groupby('UniqueCarrier')['ArrDelay'].mean().sort_values(ascending=False).head(10)
```

```python
df_sample.groupby('UniqueCarrier')['ArrDelay'].sum().sort_values(ascending=False).head(10)
```

```python
df_sample.groupby('UniqueCarrier')['ArrDelay'].count().sort_values(ascending=False).head(10)
```

```python
#Creamos una nueva columna que dice si el vuelo ha tenido o no retraso
df_sample['Delayed'] = df_sample['ArrDelay'] >= 15
df_sample.head(10)
```

```python
# Repetimos con la nueva columna
df_sample.groupby('UniqueCarrier')['Delayed'].mean().sort_values(ascending=False).head(10)
```

```python
df_sample.groupby('UniqueCarrier')['Delayed'].sum().sort_values(ascending=False).head(10)
```

```python
df_sample.groupby('UniqueCarrier')['Delayed'].count().sort_values(ascending=False).head(10)
```
