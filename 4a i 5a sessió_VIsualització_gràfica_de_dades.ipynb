```python
# Importamos para hacer plot
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Escogemos un stilo por defecto para graficar
# matplotlib.org tenemos toda la info de la librería y sus estilos
plt.style.use('classic')
```

```python
# si trabajamos solo en python para que muestre hay que dicer plt.show()
```

```python
# Definimos x como 100 valores espaciados entre 0 y 10
x = np.linspace(0,10,100)
fig = plt.figure()
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x), '.')
plt.plot(x,np.cos(x), '--')
```

```python
# Para hacer subdivisiones con subplot indicando numcol, numfil y num de panel (1 arriba, 2 abajo)
plt.subplot(2,1,1)
plt.plot(x,np.sin(x))

plt.subplot(2,1,2)
plt.plot(x,np.sin(x))
```

```python
# 2 filas y 2 columnas, en los paneles 2 y 3
plt.subplot(2,2,2)
plt.plot(x,np.sin(x))

plt.subplot(2,2,3)
plt.plot(x,np.sin(x))
```

```python
# Más ejemplos, 2 filas, la de arriba de 1 col y la de abajo de 2 columnas (y pinto en)
plt.subplot(2,1,1)
plt.plot(x,np.sin(x))

plt.subplot(2,2,4)
plt.plot(x,np.sin(x))
```

```python
# Más ejemplos (NO ESTÁ BIEN)
plt.subplot(2,1,1)
plt.plot(x,np.sin(x))

plt.subplot(2,4,3)
plt.plot(x,np.sin(x))

plt.subplot(2,2,3)
plt.plot(x,np.sin(x))
```

```python
# Orientado a Objetos
fig, ax = plt.subplots(2)
ax[0].plot(x,np.sin(x))
ax[1].plot(x,np.sin(x))
```

```python
#scatter con parámetros variados
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.randn(100)
sizes = rng.randn(100)*100
plt.scatter (x,y, c=colors, cmap='viridis', sizes= sizes, alpha=0.5)
plt.colorbar()
```
