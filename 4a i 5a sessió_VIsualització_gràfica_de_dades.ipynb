```python
#Damos un orden a los días (el que queramos) por importancia o lo que sea
days_order = ['Fri','Thur','Sun', 'Sat']
```

```python
# Crear una función anónima
def map_day_order(day):
    return days_order.index(day)
tips.day.apply(map_day_order)
```

```python
#Otra forma más sencilla; con lambda sin necesidad de crear la función
tips.day.apply(lambda day: days_order.index(day))
```

```python
#Creamos una nueva columna con ese valor
tips['days_order'] = tips.day.apply(lambda day: days_order.index(day))
tips
```

```python
# Podemos hacer un gráfico de línea
sns.lineplot(x='days_order', y= 'total_bill', data=tips)
```

```python
# la zona sombreada alrededor de la línea, es la zona de incertidumbre/error. Con muestras mayores, esta zona es menor, pero crecería
# la dispersión de los datos

# Para eliminar la zona de error
sns.lineplot(x='days_order', y= 'total_bill', data=tips, errorbar=None)
```

```python
# Errorbar con 2 desviacion estándar
sns.lineplot(x='days_order', y= 'total_bill', data=tips, errorbar=('sd',2))
```

```python
# También podemos hacerlo con percentiles
#P.e. el 95% de los datos
sns.lineplot(x='days_order', y= 'total_bill', data=tips, errorbar=('pi',1))
```

```python
#
sns.lineplot(x='days_order', y= 'total_bill', data=tips, errorbar=('pi',50))
```

```python
# Con error estándar de la media
#la media de dos errores estándar
sns.lineplot(x='days_order', y= 'total_bill', data=tips, errorbar=('se',2))
```

```python
# El intervalo de confianza 
#del 95% p.e.
sns.lineplot(x='days_order', y= 'total_bill', data=tips, errorbar=('ci',95))
```

```python
#intervalo de confianza del 10%
sns.lineplot(x='days_order', y= 'total_bill', data=tips, errorbar=('ci',10))
```

```python
#Como el matplotlib podemos añadir capas si añadimos en la misma celda
sns.lineplot(x='days_order', y= 'total_bill', data=tips)
sns.lineplot(x='days_order', y= 'total_bill', hue = 'day',data=tips)
```

```python
#Combinamos seaborn y matplotlib
sns.PairGrid(tips).map(plt.scatter)
```

```python
#En la diagonal ponemos el kdeplot; fuera de la diagonal scatter
sns.PairGrid(tips).map_diag(sns.kdeplot).map_offdiag(sns.scatterplot)
```

```python
#Diferenciamos qué queremos por encima y por debajo de la diagonal
sns.PairGrid(tips, hue='sex').map_diag(sns.kdeplot).map_upper(sns.lineplot).map_lower(sns.scatterplot)
```

```python
#Lo mismo sin desglose de sex
sns.PairGrid(tips).map_diag(sns.kdeplot).map_upper(sns.lineplot).map_lower(sns.scatterplot)
```
