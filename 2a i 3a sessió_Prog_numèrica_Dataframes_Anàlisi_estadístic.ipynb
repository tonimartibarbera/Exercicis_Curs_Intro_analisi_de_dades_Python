```python
#Datos que faltan NaN, None
#Creamos un dataframe a partir de un diccionario
dict = {'Col1':[0,1,2,np.nan,4,5,6,7], 'Col2':[1.2,5.5,7.7,np.nan,9.9,10,11,12], 'Col3':[None,'b','c','d','e','c','b','b']}
df3 = pd.DataFrame(dict)
df3
```

```python
# Demanem info del nostre dataframe
df3.info()
```

```python
# Demanem els estadísticos
df3.describe()
```

```python
# per saber quants nulls tenim en cada columna
df3.isnull().sum()
```

```python
# Al mateix amb aquesta altra funció
df3.isna().sum()
```

```python
#Per veure el nulls
df3.isnull()*1
```

```python
#Per eliminar los valores perdidos >> Sustituirlos por la media, la moda, etc.
# En la Col1 sustituimos por la media
df3['Col1'].fillna(df3['Col1'].mean(), inplace=True)
df3
```

```python
df3.isna().sum()
```

```python
#En la Col2 lo haremos por interpolación (de los 2 valores más cercanos)
df3['Col2'].interpolate(inplace=True)
df3
```

```python
# Para Col3 miramos las moda
df3.Col3.mode()
```

```python
# La guardamos en una lista
lista = df3.Col3.mode()
lista[0]
```

```python
# Sustituimos los None de la Col3 por la moda
df3['Col3'].fillna(df3.Col3.mode().iloc[0],inplace=True)
df3
```

```python
#Mapeo de datos a través de una función. Restar 2 a la edad de dataframe df2
df2
```

```python
def resta_edad(value):
    return value-2
df2.age = df2.age.map(resta_edad)
df2
```

```python
#Asignaciones >> crear o modificar varias columnas al mismo tiempo
df2
```

```python
df2 = df2.assign(experience = [10, 3, 19, 20, 1], children = [0, 1, 2, 0, 5], state = ['single', 'married','divorced','married','married'])
df2
```

```python
# Apply >> Aplica una función a una columna y sustituye su valor por el devuelto por la función
df2.age = df2.age.apply(resta_edad)
df2
```

```python
def estado_civil(value):
    if value == 'single':
        return 0
    elif value == 'married':
        return 1
    else:
        return 2
df2.state = df2.state.apply(estado_civil)
df2
```

```python
#concatenar dataframes, por filas o por columnas
#Creamos otro dataframe a partir de un diccionario
data={'age':[34],'profession':['teacher'],'experience':[4],'children':[1],'state':[2],'age_job':[3]}
df4 = pd.DataFrame(data)
df4
```
